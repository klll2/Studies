```python
from sklearn.datasets import load_breast_cancer 
# scikit-learn에 포함된 dataset 은 Bunch 객체에 저장되어 있음
cancer = load_breast_cancer() 

print("cancer dataset의 key: \n{}".format(cancer.keys()))

print(cancer['DESCR'][:5000]+ "\n...")   # DESCR 키: dataset 에 대한 간단한 설명 포함
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=42) 
# random_state=42 : 난수 생성기를 이용하여 dataset을 무작위로 섞음. 
# 난수 초기값을 주면 여러 번 실행해도 결과가 동일  
# train_test_split 의 return 값은 X_train, X_test, y_train, y_test (모두 numpy 배열)
# X_train 은 75%, X_test 25% (default)

print("X_train 의 크기: {}".format(X_train.shape))
print("y_train 의 크기: {}".format(y_train.shape))
print("X_test 의 크기: {}".format(X_test.shape))
print("y_test 의 크기: {}".format(y_test.shape))
```

```python
tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train, y_train)
```

```python
y_pred = tree.predict(X_test)
print("Test set에 대한 예측 값: \n {}".format(y_pred)) 
```

```python
print("Training set의 정확도: {:.3f}".format(tree.score(X_train, y_train)))
print("Test set의 정확도: {:.3f}".format(tree.score(X_test, y_test)))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_breast_cancer # scikit에 포함된 dataset 은 Bunch 객체에 저장되어 있음
cancer = load_breast_cancer() 
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=42)
tree = DecisionTreeClassifier(max_depth=4, random_state=0)
tree.fit(X_train, y_train)
print("Training set의 정확도: {:.3f}".format(tree.score(X_train, y_train)))
print("Test set의 정확도: {:.3f}".format(tree.score(X_test, y_test)))
```
