```python
from sklearn.model_selection import train_test_split
X_train, X_test , y_train , y_test = train_test_split(iris_dataset ['data'], iris_dataset ['target'], random_state =0)
#random_state =0 : 난수 생성기를 이용하여 dataset 을 무작위로 섞음 .
#섞지 않으면 test set 25% 의 label 은 뒤에 있기 때문에 모두 2 가 됨 .
#난수 초기값 0 을 주면 여러 번 실행해도 결과가 동일
#train_test_split 의 return 값은 X_train , X_test , y_train
```

```python
import pandas as pd
iris_dataframe = pd.DataFrame(X_train , columns=iris_dataset.feature_names)
#X_train 을 사용하여 data frame 을 만듦 .
# column name은 iris_dataset.feature_names 사용 .
```

```python
import mglearn
pd.plotting.scatter_matrix(iris_dataframe , c=y_train , figsize =(15,15), marker='o', hist_kwds ={'bins':20}, s=60, alpha=.9, cmap =mglearn.cm3)
# data frame을 사용하여 y_train 에 따라 구분된 산점도 행렬을 만듦 , 히스토그램과 산점도 그래프가 동시에 출력됨
# c=y_train : 컬러를 target attribute 인 y 에 맞게 구분
#figsize =(15,15) : graph 의 행 , 열의 크기
# marker='o' : dot종류
# alpha=.9: 투명도
#cmap =mglearn.cm3 : 시각화를 위한 컬러 매핑
#hist_kwds ={'bins':20}: 히스토그램 관련 옵션 사용 . bin 수 =
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors =1)
```

```python
knn.fit(X_train , y_train)
```

```python
KNeighborsClassifier(algorithm='auto', leaf_size =30, metric='minkowski', metric_params=None, n_jobs =None, n_neighbors =1, p=2, weights='uniform')
```

```python
import numpy as np
X_new = np.array ([[5, 2.9, 1, 0.2]])
print("X_new.shape : {}".format(X_new.shape ))
```

```python
prediction = knn.predict(X_new)
print("예측 : {}".format(prediction))
print("예측한 target 의 이름 : {}".format(iris_dataset['target_names'][prediction]))
```

```python
y_pred = knn.predict(X_test)
print("Test set에 대한 예측 값 : n {}".format( y_pred ))
```

```python
print("Test set에 대한 실제 값 : n {}".format( y_test ))
```

```python
print("Test set의 정확도 : {:.2f}".format(np.mean(y_pred==y_test)))
# : {:,2f} -> 소수점 밑 2자리수까지 출력
```

```python
print("Test set의 정확도 : {:.2f}".format(knn.score(X_test , y_test)))
```
